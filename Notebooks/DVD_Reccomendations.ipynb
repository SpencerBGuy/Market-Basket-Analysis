```python
df = pd.read_csv('https://raw.githubusercontent.com/SpencerBGuy/Market-Basket-Analysis/master/Data/DVD_POS_Txns.csv')
df.head()
```

```python
basket = (df.groupby(['PO_ID','Item_desc'])['Txns']
          .sum().unstack().reset_index().fillna(0)
          .set_index('PO_ID'))
basket.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
basket_sets
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.001, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=.0001)
rules.head()
```

```python
rules[ (rules['lift'] >= 1) &
       (rules['confidence'] >= 0.1) ]
```
